```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import re
import matplotlib as mpl

import matplotlib.font_manager as fm

mpl.rcParams['axes.unicode_minus']= False

path = 'C:/Windows/Fonts/malgun.ttf'

font_name = fm.FontProperties(fname = path, size = 50).get_name()

plt.rc('font', family=font_name)
```

```python
area = ['부산','충북','충남','대구','대전','강원','광주','경기','경북','경남','인천','제주','전북','전남','세종','서울','울산']
num = [658,321,325,322,325,597,228,803,597,419,218,117,332,408,39,1022,117]
star = [199,101,83,79,83,191,49,258,191,111,66,28,82,108,9,266,28]
```

```python
fig = plt.figure(figsize=(8,8)) ## Figure 생성 
fig.set_facecolor('white') ## Figure 배경색 지정
ax1 = fig.add_subplot() ## axes 생성

colors = sns.color_palette('summer', len(area)) ## 바 차트 색상

xtick_label_position = list(range(len(area))) ## x축 눈금 라벨이 표시될 x좌표
ax1.set_xticks(xtick_label_position) ## x축 눈금 
ax1.set_xticklabels(area) ## x축 눈금 라벨
ax1.bar(xtick_label_position, num, color=colors, label='총 업소 수') ## 바차트 출력
plt.xlabel('지역명')
plt.ylabel('업소수')
plt.legend()

color = 'blue'
ax2 = ax1.twinx() ## 새로운 axis 생성
ax2.plot(xtick_label_position, star, color=color, linestyle='--', marker='o',label = '평점4.0 이상 업소의 수') ## 선 그래프 
plt.legend()
ax2.tick_params(axis='y', labelcolor=color) ## 눈금 라벨 색상 지정
plt.ylim([0, 500])     # Y축의 범위: [ymin, ymax]

plt.title('총 착한 가격 업소 수 대비 평점 4.0 이상 업소 수', fontsize=15)
plt.show()
```
