```python
def ad_check(x):
    L = []
    if x not in L:
        L.append(x)

    return ' '.join(L)
```

```python
dd['주소'] = dd['주소'].apply(lambda x: ad_check(str(x)))
dd
```

```python
dd2 = dd1.drop(dd1[(dd1['카카오별점'] == '별점없음') & (dd['네이버별점'] == '별점없음')].index)
dd2.위도 = dd2.위도.astype('float')
lat = dd2.위도.mean()
long = dd2.경도.mean()

#map객체 생성
m = folium.Map([lat, long])

marker_cluster = MarkerCluster().add_to(m)

#map에 marker를 추가
for i in dd2.index:
  sub_lat = dd2.loc[i, '위도']
  sub_long = dd2.loc[i, '경도']
  title = dd2.loc[i,'주소']
  title2 = dd2.loc[i, '업종'] + r' | ' + dd2.loc[i, '업소명'] + r' | ' +'카카오별점 = ' + dd2.loc[i, '카카오별점'] + r' | ' +'네이버별점 = ' + dd2.loc[i, '네이버별점']
  icon_color ='blue'
  folium.Marker([sub_lat, sub_long],
                icon=folium.Icon(color='orange',
        icon_color='white',
        icon='glyphicon glyphicon-thumbs-up', # googledp glyphicon 검색해서 관련된 코드 제공해주는 사이트들 참고해서 원하는 것 가져오기
        prefix='glyphicon'),
                popup=f'<i>{title}</i>',
                tooltip = title2).add_to(marker_cluster)

m
```
